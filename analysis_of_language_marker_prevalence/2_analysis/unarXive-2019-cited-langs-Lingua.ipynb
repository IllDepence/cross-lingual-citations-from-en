```python
# Generate scaffold for manually checking 392 papers for untranslated Russian references
#                                      and 30 papers for untranslated Ukranian references
#                                       and 6 papers for untranslated Bulgarian references

import sqlite3
import re
con = sqlite3.connect('unarXive-2020/papers/refs.db')
cur = con.cursor()
fn_patt = re.compile('([A-Za-z-]+)(.+)')
```

```python
bulgarian_aids_to_check = []
for uuid in df_refsecitems_2019[df_refsecitems_2019.lingua_language == 'BULGARIAN'].uuid.to_list():
    tpl = cur.execute(
        'select citing_arxiv_id from bibitem where uuid = "%s"' % uuid
    ).fetchone()
    bulgarian_aids_to_check.append(tpl[0])
    
for i, aid in enumerate(list(set(bulgarian_aids_to_check))):
    m = fn_patt.search(aid)
    if m:
        aid = '{}/{}'.format(m.group(1), m.group(2))
    print('[] [{0}] <a href="http://arxiv.org/pdf/{1}">http://arxiv.org/pdf/{1}</a><br>'.format(i, aid))
    
# Manual check:
# 0 untranslated Bulgarian references (in 0 documents)
```

```python
ukranian_aids_to_check = []
for uuid in df_refsecitems_2019[df_refsecitems_2019.lingua_language == 'UKRAINIAN'].uuid.to_list():
    tpl = cur.execute(
        'select citing_arxiv_id from bibitem where uuid = "%s"' % uuid
    ).fetchone()
    ukranian_aids_to_check.append(tpl[0])
    
for i, aid in enumerate(list(set(ukranian_aids_to_check))):
    m = fn_patt.search(aid)
    if m:
        aid = '{}/{}'.format(m.group(1), m.group(2))
    print('[] [{0}] <a href="http://arxiv.org/pdf/{1}">http://arxiv.org/pdf/{1}</a><br>'.format(i, aid))
    
# Manual check:
# 15 untranslated Ukrainian references in 4 documents (arxiv.org/pdf/1412.5859, arxiv.org/pdf/1412.5305,
#                                                      arxiv.org/pdf/1102.4250, arxiv.org/pdf/1207.3618,)
```
